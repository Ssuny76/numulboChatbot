```python
html = driver.page_source
soup = BeautifulSoup(html, "lxml")
item_images = soup.select('div[class=prodListWrap] > ul > li > div[class="photo"] > a > img')
item_names = soup.select('div[class="prodListWrap"] > ul > li > p[class="prodName"] > span')
bigGroup = soup.select('h1')

result = []
for i in range(len(item_names)):
    item_dict = {}
    item_dict['category'] = bigGroup[0].text.strip()
    item_dict['item_name'] = item_names[i].text
    item_dict['img_src'] = item_images[i]['src']
    result.append(item_dict)
    print(item_dict)
```

```python
exists = os.path.isfile('./result_src.csv')

if exists:
    with open('./result_src.csv', 'a') as csvfile:
        fieldnames = ["category", "item_name", "img_src"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for i in range(len(result)):
            writer.writerow(result[i])
else:
    with open('./result_src.csv', 'w') as csvfile:
        fieldnames = ["category", "item_name", "img_src"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for i in range(len(result)):
            writer.writerow(result[i])
```
